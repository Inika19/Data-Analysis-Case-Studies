```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# sns.set()
# sns.set_style('dark')
```

```python
df = pd.read_csv('train.csv', index_col='PassengerId')
test_df = pd.read_csv('test.csv', index_col='PassengerId')
```

```python
df.head()
```

```python
sns.catplot(data = df, x = 'Survived', kind='count')
plt.xticks([0,1],['No','Yes'])
plt.title("Survivors Count",fontsize=12)
plt.show()
```

```python
d = {'C' : 'Cherbourg', 'Q' : 'Queenstown', 'S' : 'Southampton'}
_ = plt.figure(figsize = (12,6))
g = sns.catplot(data = df, x = 'Survived', col = 'Embarked', kind = 'count')
plt.xticks([0,1], ['No','Yes'])
g.fig.suptitle("Port Embarked from vs Survival",y = 1.1, fontsize = 14)
for i in range(3):
    key = g.axes[0][i].title.get_text()[-1]
    g.axes[0][i].set_title('Embarked from: {}'.format(d[key]))
plt.show()
```

```python
plt.figure(figsize=(12,8))
g = sns.catplot(data = df, x = 'Survived', kind = 'count', col = 'Pclass')
g.fig.suptitle('Survivors by Passenger Class',y=1.1,fontsize=15)
g.set_titles('Passenger Class : {col_name}',fontsize=14)
_ = plt.xticks([0,1],['No','Yes'])
plt.show()
print("Number of deaths was greater among lower class passengers")
```

```python
plt.figure(figsize=(12,8))
g = sns.catplot(data = df, x = 'Survived', kind = 'count', col = 'Sex')
g.fig.suptitle('Survivors by Sex',y=1.1,fontsize=15)
g.set_titles('{col_name}',fontsize=14)
_ = plt.xticks([0,1],['No','Yes'])
plt.show()
print("More women than men survived")
```

```python
plt.figure(figsize = (12,8))
g = sns.catplot(data = df,x = 'Survived', y = 'Age', kind = 'box')
g.fig.suptitle('Age vs Survival',y=1.1,fontsize=15)
plt.xticks([0,1],['No','Yes'])
plt.show()
print("Median Age is the same in both groups")
print("Survivors were younger than those that died")
```

```python
plt.figure(figsize = (12,8))
g = sns.catplot(data = df,x = 'Survived', y = 'Age', kind = 'box',col='Sex')
g.fig.suptitle('Age vs Survival by Sex',y=1.1,fontsize=15)
g.set_titles('{col_name}')
plt.xticks([0,1],['No','Yes'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
g = sns.catplot(kind = 'point', x = 'Sex', y = 'Age', data = df, hue = 'Survived')
# plt.legend(title='Survived', loc='upper left', labels=['No', 'Yes'])
plt.show(g)
print("Of the men, the older died and the younger survived")
print("This trend was reversed in women")
```

```python
sns.swarmplot(data = df, x = 'Survived', y = 'SibSp')
plt.xticks([0,1],['No','Yes'])
plt.ylabel("No of Siblings/ Spouses Onboard")
plt.show()
```

```python
sns.swarmplot(data = df, x = 'Survived', y = 'Parch')
plt.xticks([0,1],['No','Yes'])
plt.ylabel("No of Parents/ children Onboard")
plt.show()
```

```python
sns.swarmplot(x = df['Survived'], y = df['SibSp']+df['Parch'])
plt.xticks([0,1],['No','Yes'])
plt.ylabel("Family members onboard")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6),sharex = True)
g = sns.swarmplot(data = df, x = 'Survived', y = 'Fare', ax = ax[0])
g.set(ylabel = 'Fare Paid')
g = sns.boxplot(data = df, x = 'Survived', y = 'Fare', sym = '', ax = ax[1])
g.set(ylabel = 'Fare Paid')
plt.suptitle('Fare vs Survival')
plt.xticks([0,1],['No','Yes'])
plt.show()
print("Higher the fare paid, higher was the chance of survival")
```
